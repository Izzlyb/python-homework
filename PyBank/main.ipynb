```python
# main.ipynb for PyBank homework project from Module 2 Challenge

from pathlib import Path
import csv

# Set the file path
csvpath = Path('./Resources/budget_data.csv')

# Initialize list of records
records = []
dt_line = []
variations = []
total_months = 0
total_balance = 0
max_increase = 0
mo_max_increase = ''
max_decrease = 0
mo_max_decrease = ''

# Open the csv file as an object
with open(csvpath, 'r') as csvfile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile)

    # Read the header row first (skip this step if there is no header)
    csv_header = next(csvreader)

    # @TODO: Read the header row
    # print(f"the headers are: {csv_header}\n")

    # Read the complete file and format each line with string and int data
    # add it to a list.
    for lines in csvreader:
        dt_line.append(lines[0])
        total_balance += int(lines[1])
        dt_line.append(int(lines[1]))
        records.append(dt_line)
        total_months += total_months

        dt_line = []

    average_variation = 0
    
    # Search for max variation increase or variation decrease
    for idx in range(0, len(records) - 1):
        variation = records[idx + 1][1] - records[idx][1]
        variations.append(variation)
        
        if variation > max_increase:
            max_increase = variation
            mo_max_increase = records[idx+1][0]
            # print(f"max_increase was on: {records[idx][0]} amount: {max_increase}")
            
        if variation < max_decrease:
            max_decrease = variation
            mo_max_decrease = records[idx+1][0]
            # print(f"max_decrease was on: {records[idx][0]} amount: {max_decrease}")

    average_variation = sum(variations)/len(variations)
    # print(f" The averate variation was: {average_variation}\n")
    
    output_lst = []
    line = "\nFinancial Analysis"
    output_lst.append(line)
    line = "----------------------------------------\n"
    output_lst.append(line)    
    
    # output report to screen
    print(f"\n{'Financial Analysis' : ^76}")
    print("----------------------------------------------------------------------------\n")
    print("{0:30} {1:13,d}".format("Total Months:", len(records)))
    print("{0:30} ${1:12,d}".format("Total Balance:", total_balance))
    print("{0:30} ${1:12,.2f}".format("Average change:", average_variation))
    print(f"Greatest Increase in Profit was on: {mo_max_increase} in the amount of ${max_increase:12,d}")
    print(f"Greatest Decrease in Profit was on: {mo_max_decrease} in the amount of ${max_decrease:12,d}")
    
    # output to file
    f_output = open("fin_analysis_results.txt", "w")
    f_output.write(f"\n{'Financial Analysis' : ^76}")
    f_output.write(f"\n----------------------------------------------------------------------------\n")
    f_output.write(f"\nTotal Months:        {len(records):12,d}")
    f_output.write(f"\nTotal Balance:       {total_balance:12,d}")
    f_output.write(f"\nAverage change:      {average_variation:12,.2f}")
    f_output.write(f"\nGreatest Increase in Profit was on: {mo_max_increase} in the amount of ${max_increase:12,d}")
    f_output.write(f"\nGreatest Decrease in Profit was on: {mo_max_decrease} in the amount of ${max_decrease:12,d}")
    f_output.write("\n")
    
    f_output.close()
```
